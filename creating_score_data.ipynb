```python
from music21 import *
from fractions import Fraction
```

```python
s = converter.parse("C:\\Users\\jonathansp\\Downloads\Beethoven_Symphony_No._1_in_C_Op._21_Full_Score.mxl")
```

```python
parts = []
for part in s.parts.measures(187,191):
    print(part)
    parts.append(part)
```

```python
def is_the_beat_divided(note):
    if note.quarterLength >= 1:
        return False
    else:
        
        return Fraction(note.quarterLength).denominator
        
```

```python
notes_and_inst = []
measures_and_beats = []

measures = []
beats_per_measure = []
notes_and_rests_per_inst = []

beat_cells = []

cached = []

the_notes = []
the_inst = []
    

for part in parts:
    for measure in part.getElementsByClass('Measure'):
#             print(note.measureNumber, note.nameWithOctave, note.beat, note.duration, note.getInstrument().partAbbreviation)
       
        for beat in range(int(measure.barDuration.quarterLength)):
            beats_per_measure.append(beat+1)
            measures.append(measure.measureNumber)
            
    measures_and_beats.append({part.getInstrument().partAbbreviation: measures})
    notes = []
    insts = []
    inst = ""
    for note in part.flatten().getElementsByClass(['Note']):

        if {note.measureNumber: int(note.beat)} not in cached:
            to_cache = {note.measureNumber: int(note.beat)}
            cached.append(to_cache)
            cells = is_the_beat_divided(note)
            if cells:
                data = {int(note.beat): [beat+1 for beat in range(cells)]}
                beat_cells.append({note.measureNumber: data, 'inst': note.getInstrument().partAbbreviation})
        else:
            data = {int(note.beat): 1}
            beat_cells.append({note.measureNumber: data, 'inst': note.getInstrument})
        if not inst:
            inst = note.getInstrument().partAbbreviation
        notes.append(note.nameWithOctave)
        insts.append(inst)
    the_notes.append(notes)
    the_inst.append(insts)
    notes_and_inst.append({inst: notes})
    
    print(the_notes)
print(len(the_notes), len(the_inst))

```
